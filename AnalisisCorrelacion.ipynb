# Análisis de correlación entre duración de trayectos y diferencia de tiempo

## Cálculo de media, varianza y covarianza para datos agrupados

Cuando los datos se presentan en **intervalos de frecuencia**, como en la tabla de `DRIVE`, no conocemos los valores individuales de cada observación. Para aproximarlos, se asigna a cada intervalo un **punto medio** \(x_i\), que representa todos los valores del rango de ese intervalo.

### Puntos medios de los intervalos

Para un intervalo \([a, b)\), el punto medio se calcula como:

$$
x_i = \frac{a + b}{2}
$$

- Ejemplo para los intervalos de `DRIVE`:

| Intervalo     | Punto medio \(x_i\) |
|---------------|-------------------|
| [0, 10)       | 5                 |
| [10, 20)      | 15                |
| [20, 30)      | 25                |
| [30, 45)      | 37.5              |
| [45, 60)      | 52.5              |
| [60, 90)      | 75                |
| [90, 120)     | 105               |

---

### Media para datos agrupados

La **media** se aproxima usando los puntos medios y las frecuencias absolutas \(n_i\) de cada intervalo:

$$
\bar{x} = \frac{\sum (x_i \cdot n_i)}{N}
$$

- \(x_i\) = punto medio del intervalo  
- \(n_i\) = frecuencia absoluta del intervalo  
- \(N = \sum n_i\) = total de observaciones  

---

### Varianza para datos agrupados

La **varianza** se aproxima como:

$$
\sigma^2 = \frac{\sum n_i (x_i - \bar{x})^2}{N}
$$

- Multiplicamos la frecuencia de cada intervalo por el cuadrado de la diferencia entre su punto medio y la media, y dividimos entre el total de datos.  
- La desviación estándar se obtiene como \(\sigma = \sqrt{\sigma^2}\).

---

### Covarianza para datos agrupados

Si tenemos dos variables agrupadas (por ejemplo `DRIVE` y `TRANSIT`) con puntos medios \(x_i\) y \(y_i\) y frecuencias \(n_i\), la **covarianza** se aproxima con:

$$
\text{cov}(X,Y) = \frac{\sum n_i (x_i - \bar{x})(y_i - \bar{y})}{N}
$$

- \(x_i, y_i\) = puntos medios de cada intervalo de las variables X e Y  
- \(n_i\) = frecuencia de cada intervalo (si son frecuencias conjuntas, mejor)  
- \(\bar{x}, \bar{y}\) = medias de X e Y  

Esta metodología permite calcular **media, varianza y covarianza usando datos agrupados**, lo cual es útil cuando solo se dispone de tablas de frecuencia.

En esta sección se analiza la **relación entre la duración de los trayectos** realizados en coche (`DRIVE`) o en transporte público (`TRANSIT`) y la **diferencia de tiempo entre ambos modos** (`TRANSIT - DRIVE`).

Se utilizan los siguientes estadísticos conocidos para simular los datos:

| Variable         | Media (μ) | Desviación estándar (σ) |
|-----------------|-----------|------------------------|
| DRIVE           | 64.62 min | 87.34 min              |
| TRANSIT         | 72.67 min | 75.02 min              |
| DIFERENCIA      | 8.05 min  | 54.55 min              |
| Tamaño de muestra | n = 1391 | —                      |

---

## Generación de datos simulados

Se generan **distribuciones normales** para cada modo de transporte usando las medias y desviaciones estándar conocidas:

$$
drive \sim \mathcal{N}(\mu_{\text{drive}}, \sigma_{\text{drive}}^2)
$$

$$
transit \sim \mathcal{N}(\mu_{\text{transit}}, \sigma_{\text{transit}}^2)
$$

Luego se calcula la **diferencia** entre los tiempos de transporte público y coche:

$$
diferencia = transit - drive
$$

Esta diferencia permite evaluar **cómo varía el tiempo de transporte entre ambos modos**.

---

## Coeficientes de correlación de Pearson

Para medir la **relación lineal** entre cada modo de transporte y la diferencia de tiempo, se calcula el **coeficiente de correlación de Pearson** (`r`):

$$
r_{X,Y} = \frac{\text{cov}(X,Y)}{\sigma_X \sigma_Y}
$$

- `X` es la duración del trayecto (`DRIVE` o `TRANSIT`)  
- `Y` es la diferencia de tiempo (`TRANSIT - DRIVE`)  
- `cov(X,Y)` es la covarianza entre X y Y  
- `σ_X`, `σ_Y` son las desviaciones estándar de X y Y  

Además, se calcula **r²**, que representa el **porcentaje de variabilidad en Y explicado por X**:

$$
r^2 = (r_{X,Y})^2
$$

### Resultados obtenidos

| Comparación             | r      | r²    |
|-------------------------|--------|-------|
| DRIVE vs DIFERENCIA     | -0.761 | 0.579 |
| TRANSIT vs DIFERENCIA   | 0.689  | 0.475 |

### Interpretación

- La **correlación negativa** entre `DRIVE` y la diferencia (`r = -0.761`) indica que **a mayor duración en coche, menor es la diferencia** con respecto al transporte público.  
  Es decir, trayectos largos en coche tienden a reducir la diferencia de tiempo.  

- La **correlación positiva** entre `TRANSIT` y la diferencia (`r = 0.689`) sugiere que **a mayor duración en transporte público, mayor es la diferencia** con respecto al coche.  

En términos de `r²`:

- Aproximadamente **57.9% de la variabilidad de la diferencia** puede explicarse por la duración en coche.  
- Aproximadamente **47.5% de la variabilidad** puede explicarse por la duración en transporte público.  

Esto indica que la **duración del trayecto en coche tiene un impacto ligeramente mayor sobre la diferencia de tiempo** que la duración en transporte público, aunque ambos factores son relevantes.

## Representación gráfica de la relación duración vs diferencia

Se generan **gráficos de dispersión** para visualizar la relación entre los tiempos de trayecto y la diferencia (`TRANSIT - DRIVE`).

### Gráfico DRIVE vs DIFERENCIA

- **Eje X**: duración del trayecto en coche (`DRIVE`)

**Interpretación visual:**

---

## Conclusión general

A partir del análisis de correlación y los gráficos:

- La **duración del trayecto en coche** está fuertemente relacionada con la **reducción de la diferencia de tiempo** con respecto al transporte público.  
- La **duración del trayecto en transporte público** se relaciona positivamente con la diferencia, indicando que trayectos largos en transporte público **aumentan la diferencia frente al coche**.  
- Los **gráficos de dispersión confirman visualmente estas tendencias**, mostrando patrones claros de relación lineal que coinciden con los coeficientes de Pearson obtenidos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Estadísticos conocidos
# -----------------------------
mean_drive = 64.623652
std_drive = 87.336185

mean_transit = 72.673113
std_transit = 75.018465

mean_diff = 8.049461
std_diff = 54.548856

n = 1391

# -----------------------------
# Generar datos simulados
# -----------------------------
np.random.seed(1)

drive = np.random.normal(mean_drive, std_drive, n)
transit = np.random.normal(mean_transit, std_transit, n)

diferencia = transit - drive

df = pd.DataFrame({
    "drive": drive,
    "transit": transit,
    "diferencia": diferencia
})

# -----------------------------
# Correlaciones de Pearson
# -----------------------------
r_drive = df["drive"].corr(df["diferencia"])
r_transit = df["transit"].corr(df["diferencia"])

r2_drive = r_drive**2
r2_transit = r_transit**2

print("\nCoeficientes de correlación de Pearson\n")

print(f"DRIVE vs DIFERENCIA (r): {r_drive:.3f}")
print(f"DRIVE vs DIFERENCIA (r²): {r2_drive:.3f}\n")

print(f"TRANSIT vs DIFERENCIA (r): {r_transit:.3f}")
print(f"TRANSIT vs DIFERENCIA (r²): {r2_transit:.3f}\n")

# -----------------------------
# Gráfico DRIVE vs DIFERENCIA
# -----------------------------
plt.figure()

plt.scatter(df["drive"], df["diferencia"], alpha=0.5)

plt.xlabel("Duración DRIVE (min)")
plt.ylabel("Diferencia (TRANSIT - DRIVE)")
plt.title("Relación entre duración en coche y diferencia de tiempo")

plt.grid(True)

plt.show()

# -----------------------------
# Gráfico TRANSIT vs DIFERENCIA
# -----------------------------
plt.figure()

plt.scatter(df["transit"], df["diferencia"], alpha=0.5)

plt.xlabel("Duración TRANSIT (min)")
plt.ylabel("Diferencia (TRANSIT - DRIVE)")
plt.title("Relación entre duración en transporte público y diferencia")

plt.grid(True)

plt.show()
```
